```python
print("part (f)\n")

# Reload the dataset if not already loaded
df = pd.read_csv('C:\\Users\\Dell\\Desktop\\Machine learnng\\week 3\\automobile\\automobile\\dataset.csv', header=None)

# Drop duplicated rows
df_new = df.drop_duplicates()

# Save or display the unique rows DataFrame
print(df_new)  # To display in console
df_new.to_csv("f.) unique_rows_dataset.csv", index=False)  # To save to a file
```

```python
print("part (g)\n")

# Drop rows with missing values from df_new
df_new_dropped = df_new.dropna()

# Find the number of remaining rows
remaining_rows = df_new_dropped.shape[0]

print("Number of remaining rows:",remaining_rows)
```

```python
print("part (h) i.)\n")

# Fill missing values for numerical columns with mean and for string columns with the most frequent value
df_filled = df_new.copy()

for column in df_filled.columns:
    if df_filled[column].dtype in ['float64', 'int64']:
        # Replace missing numerical values with mean
        df_filled[column].fillna(df_filled[column].mean(), inplace=True)
    else:
        # Replace missing string values with the most frequent value
        df_filled[column].fillna(df_filled[column].mode()[0], inplace=True)
        
df_filled.to_csv('(h) i.) filled_dataset.csv', index=False)
df_filled
```

```python
print("part (h) ii.)\n")

# Fill missing values for numerical columns with median and for string columns with the most frequent value
df_filled_median = df_new.copy()

for column in df_filled_median.columns:
    if df_filled_median[column].dtype in ['float64', 'int64']:
        # Replace missing numerical values with median
        df_filled_median[column].fillna(df_filled_median[column].median(), inplace=True)
    else:
        # Replace missing string values with the most frequent value
        df_filled_median[column].fillna(df_filled_median[column].mode()[0], inplace=True)

# Save or display the filled DataFrame
df_filled_median.to_csv('(h) ii.) filled_with_median_and_mode.csv', index=False)
df_filled_median.head()
```

```python
print("part (j)\n")

# Identify categorical columns in the DataFrame
categorical_columns = df_new.select_dtypes(include=['object']).columns.tolist()

# Perform one-hot encoding for the identified categorical columns
df_one_hot_encoded = pd.get_dummies(df_new, columns=categorical_columns, drop_first=True)

# Extract the newly created one-hot encoded columns
encoded_columns = [col for col in df_one_hot_encoded.columns if col not in df_new.columns]

# Display the transformed (one-hot encoded) columns
encoded_data = df_one_hot_encoded[encoded_columns]

print(encoded_data.head())
encoded_data.to_csv("j.) one_hot_encoded_columns.csv", index=False)
```
